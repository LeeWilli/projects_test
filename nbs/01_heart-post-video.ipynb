```python
import requests
import threading
import multiprocessing
import time

#url = "http://localhost:8293/process_video"

# 上传的文件和参数
best = True
data = {'camera_fps': 25, 'best': best, 'save_video': False}

def make_request():
    files = {'video': open(video_path, 'rb')}
    response = requests.post(hr_server_url, files=files, data=data)
    
    if best is False:
        results = response.json()
        for r in results:
            print(r)
    else:
        print(response.text)

#make_request()

while True:
    start_time = time.time()
    # 创建10个进程并启动
    processes = []
    for _ in range(10):
        process = multiprocessing.Process(target=make_request)
        process.start()
        processes.append(process)

    # 等待所有进程执行完成
    for process in processes:
        process.join()

    end_time = time.time()
    # 计算耗时时间
    elapsed_time = end_time - start_time

    print(f"10个进程执行耗时: {elapsed_time} 秒")

    #break
```
